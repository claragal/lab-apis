```python
# Your answer here

import requests

# Function to get top tracks of an artist
def get_top_tracks(artist_name):
    # Search for the artist by name
    result = sp.search(q=f"artist:{artist_name}", type="artist")
    
    # Check if the artist is found
    if result['artists']['items']:
        # Get the artist's Spotify ID
        artist_id = result['artists']['items'][0]['id']
        
        # Get the top tracks for the artist (from the artist's Spotify ID)
        top_tracks = sp.artist_top_tracks(artist_id)
        
        # Extract the names of the top 5 tracks
        track_names = [track['name'] for track in top_tracks['tracks'][:5]]
        return track_names
    else:
        print(f"Artist {artist_name} not found on Spotify.")
        return []

# List of favorite artists
artists = ["Nick Cave", "Interpol", "Buena Vista Social Club"]

# List to store top tracks for each artist
top_tracks_list = []

# Loop through each artist and get their top tracks
for artist in artists:
    top_tracks = get_top_tracks(artist)
    top_tracks_list.append({artist: top_tracks})

# Print the result
for artist_tracks in top_tracks_list:
    print(artist_tracks)
```

```python
# Function to get related artists for an artist
def find_related_artists(artist_name):
    # Search for the artist by name
    result = sp.search(q=f"artist:{artist_name}", type="artist")
    
    # Check if the artist is found
    if result['artists']['items']:
        # Get the artist's Spotify ID
        artist_id = result['artists']['items'][0]['id']
        
        # Get related artists
        related_artists = sp.artist_related_artists(artist_id)
        
        # Extract the names of the first 5 related artists
        related_artist_names = [artist['name'] for artist in related_artists['artists'][:5]]
        return related_artist_names
    else:
        print(f"Artist {artist_name} not found on Spotify.")
        return []

# List to store related artists for each artist
related_artists_list = []

# Loop through each artist and find their related artists
for artist in artists:
    related_artists = find_related_artists(artist)
    related_artists_list.append({artist: related_artists})

# Print the result
for related in related_artists_list:
    print(related)
```

```python
# Function to get top tracks of an artist
def get_top_tracks(artist_name):
    result = sp.search(q=f"artist:{artist_name}", type="artist")
    if result['artists']['items']:
        artist_id = result['artists']['items'][0]['id']
        top_tracks = sp.artist_top_tracks(artist_id)
        track_info = [(track['name'], artist_name) for track in top_tracks['tracks'][:5]]
        return track_info
    else:
        print(f"Artist {artist_name} not found on Spotify.")
        return []

# Function to get related artists for an artist
def find_related_artists(artist_name):
    result = sp.search(q=f"artist:{artist_name}", type="artist")
    if result['artists']['items']:
        artist_id = result['artists']['items'][0]['id']
        related_artists = sp.artist_related_artists(artist_id)
        related_artist_names = [artist['name'] for artist in related_artists['artists'][:5]]
        return related_artist_names
    else:
        print(f"Artist {artist_name} not found on Spotify.")
        return []

# Function to get top tracks for related artists
def get_related_artists_top_tracks(related_artists):
    all_related_tracks = []
    for artist in related_artists:
        tracks = get_top_tracks(artist)
        all_related_tracks.extend(tracks)
    return all_related_tracks

# List of favorite artists
artists = ["Nick Cave", "Interpol", "Buena Vista Social Club"]

# List to store the final playlist tracks
playlist_tracks = []

# Loop through each artist to get their top tracks and related artists' top tracks
for artist in artists:
    # Get top tracks of the favorite artist
    top_tracks = get_top_tracks(artist)
    playlist_tracks.extend(top_tracks)
    
    # Get related artists and their top tracks
    related_artists = find_related_artists(artist)
    related_tracks = get_related_artists_top_tracks(related_artists)
    playlist_tracks.extend(related_tracks)

# Remove duplicates by converting to a set of tuples
playlist_tracks = list(set(playlist_tracks))

# Print the final playlist tracks with artist names
print("Playlist Tracks:")
for track, artist in playlist_tracks:
    print(f"{track} by {artist}")
```

### Getting Artists from a Playlist

To extract all the artists from the tracks in a playlist, you'd typically follow these steps:

1. Fetch the playlist's tracks.
2. Iterate through each track.
3. For each track, extract the associated artists.

```python
# List to store unique artist names
artists_list = []

for track_item in tracks['items']:
    track = track_item['track']
    for artist in track['artists']:
        artist_name = artist['name']
        if artist_name not in artists_list:  # This ensures each artist is added only once
            artists_list.append(artist_name)

print(artists_list)
```
